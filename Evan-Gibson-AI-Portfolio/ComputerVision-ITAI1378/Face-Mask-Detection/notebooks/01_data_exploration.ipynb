# 01 — Data Exploration
Explore the face mask dataset: class distribution, sample images, and image properties.

```python
import kagglehub
import os

path = kagglehub.dataset_download('vijaykumar1799/face-mask-detection')
print(f'Downloaded to: {path}')

# Find class folders
for root, dirs, files in os.walk(path):
    level = root.replace(path, '').count(os.sep)
    if level <= 2:
        indent = ' ' * 2 * level
        print(f'{indent}{os.path.basename(root)}/')
        if not dirs:
            print(f'{indent}  {len(files)} files')
```

## Class Distribution
Check that classes are balanced.

```python
from torchvision import datasets
import matplotlib.pyplot as plt

# Update DATA_DIR to match the printed path above
DATA_DIR = path  # adjust if class folders are in a subfolder

dataset = datasets.ImageFolder(DATA_DIR)
print(f'Total images: {len(dataset)}')
print(f'Classes: {dataset.classes}')

counts = {c: 0 for c in dataset.classes}
for _, label in dataset:
    counts[dataset.classes[label]] += 1

plt.bar(counts.keys(), counts.values(), color=['steelblue', 'seagreen', 'tomato'])
plt.title('Class Distribution')
plt.ylabel('Image Count')
plt.tight_layout()
plt.savefig('class_distribution.png', dpi=150)
plt.show()
print(counts)
```

## Sample Images
Display a sample from each class.

```python
from PIL import Image
import os

fig, axes = plt.subplots(3, 5, figsize=(15, 9))

for row, cls in enumerate(dataset.classes):
    cls_dir = os.path.join(DATA_DIR, cls)
    imgs = os.listdir(cls_dir)[:5]
    for col, img_file in enumerate(imgs):
        img = Image.open(os.path.join(cls_dir, img_file)).convert('RGB')
        axes[row][col].imshow(img)
        axes[row][col].axis('off')
        if col == 0:
            axes[row][col].set_title(cls, fontsize=9, loc='left')

plt.suptitle('Sample Images per Class', fontsize=13)
plt.tight_layout()
plt.savefig('sample_images.png', dpi=150)
plt.show()
```

## Image Size Distribution
Check consistency of image dimensions.

```python
widths, heights = [], []
sample_size = 200

for cls in dataset.classes:
    cls_dir = os.path.join(DATA_DIR, cls)
    for fname in os.listdir(cls_dir)[:sample_size // 3]:
        img = Image.open(os.path.join(cls_dir, fname))
        w, h = img.size
        widths.append(w)
        heights.append(h)

print(f'Width  — min: {min(widths)}, max: {max(widths)}, avg: {sum(widths)/len(widths):.0f}')
print(f'Height — min: {min(heights)}, max: {max(heights)}, avg: {sum(heights)/len(heights):.0f}')
print('All images will be resized to 224x224 during training.')
```
